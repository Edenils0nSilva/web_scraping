```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import pyautogui
import pandas as pd


navegador = webdriver.Chrome()

navegador.get("https://www.google.com.br/maps/@-5.1952368,-37.34164,13z?entry=ttu")
```

```python
from selenium.webdriver.common.keys import Keys

navegador.find_element('xpath', '//*[@id="searchboxinput"]').click()
navegador.find_element('xpath', '//*[@id="searchboxinput"]').send_keys("academia")
navegador.find_element('xpath', '//*[@id="searchboxinput"]').send_keys(Keys.ENTER)

time.sleep(10)
```

```python

# Encontra os elementos que representam os estabelecimentos

lojas = navegador.find_elements(By.CLASS_NAME, 'qBF1Pd')

# Imprime o número de estabelecimentos encontrados
print(f"Total de lojas de roupas encontradas: {len(lojas)}")

#criando uma lista pra guardar as informações
nomes_lojas = []

#Itera sobre as lojas e imprime seus nomes
for loja in lojas:
    nome_da_loja = loja.text
    nomes_lojas.append(nome_da_loja)
print(loja.text)

# Fechar o navegador quando terminar
#navegador.quit()

```

```python


dados = pd.DataFrame({'Nome da Loja': nomes_lojas})

pd.set_option('display.max_rows', 120)
pd.set_option('display.max_columns', 1)

display(dados)
```
